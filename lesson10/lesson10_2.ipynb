```python
#建立學生分數
#key:value
from random import randint

stu1:dict[str,int] = {
    '國文':randint(50,100),
    '英文':randint(50,100),
    '數學':randint(50,100),
    '物理':randint(50,100),
    '人文':randint(50,100)
    }
stu1
```

```python
#for...in + list

stu2:dict[str,int] = {} #建立空的dictionary

for key in ['國文','英文','數學','物理','人文']:
    stu2[key] = randint(50,100)
stu2
```

```python
#dict的comprehension -> {key:value for...in...}
#comprehesion為快速建立

stu3:dict[str,int] = {key:randint(50,100) for key in['國文','英文','數學','物理','人文']}
stu3
```

```python
#50個學生，5個分數
#dict不會依照順序，是透過key抓資料
from pprint import pprint

students:list[dict[str,int]] = [] #建立空的list
for _ in range(50):
    stu:dict[str,int] = {key:randint(50,100) for key in['國文','英文','數學','物理','人文']}
    students.append(stu)
pprint(students)
```

```python
#只有list有順序
for student in students:
    print(student)
```

```python
for student in students:
    for key in['英文','物理','國文']:
        print(f'{key}={student[key]}')
    print("=========")
```
