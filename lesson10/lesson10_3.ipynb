```python
file = open('names.txt',encoding='utf-8')
#處理file實體
file.close() #實體名稱+實體方法
if file.closed:  #closed為實體屬性(儲存值)
    print('file物件被關閉')
else:
    print('file物件沒有關閉')
```

```python
import random
def get_names(nums=1) ->list[str]:
    with open('names.txt',encoding='utf-8') as file:
        names:str = file.read()
    name_list: list[str] = names.split(sep='\n')
    return random.choices(name_list,k=nums)
```

```python
get_names(5)
```

```python
#建立50個學生名字
names:list[str] = get_names(50)
names
```

```python
from random import randint
names:list[str] = get_names(50)
students:list[dict] = []
for name in names:
    student:dict = {'姓名':name}
    for subject in ['國文','英文','數學','地理','生物']:
        student [subject] = randint(50,100)
    students.append(student)

students
```
