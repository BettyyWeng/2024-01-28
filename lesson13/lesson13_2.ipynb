```python
import requests
url_str = 'https://data.moenv.gov.tw/api/v2/aqx_p_488?api_key=e8dd42e6-9b8b-43f8-991e-b3dee723a52d&limit=1000&sort=datacreationdate desc&format=JSON'
res = requests.get(url_str) #res=response的實體
res.status_code 
```

```python
#datetime為內建函式
#第一個datetime為module，第二個datetime為class
#先建裡面的class，再建外面的class
#BaseModel專門用在dict
from pydantic import BaseModel
from datetime import datetime

class Site(BaseModel):
    sitename:str
    county:str
    aqi:int
    status:str
    datacreationdate:datetime

class Root(BaseModel):
    records:list[Site]
```

```python
#建立實體
#class的方法
root = Root.model_validate(res.json())
root
```

```python
root.records
```

```python
for site in root.records:
    print(site)
```
