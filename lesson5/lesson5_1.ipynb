```python
#學生總分為300
#有些學生可以加5%
#如果加分超過300分,就是300分

#type hint (只在建立變數時寫)
#資料型別的提醒(可有可無，有比較好)

#多個條件,使用巢項判斷

scores:int = int(input("請輸入學生分數(最高300分):"))
if scores <=300:
    is_add:str = input("學生是否符合加分條件?(y,n):")
    if is_add == "y":
        scores *=1.05
        if scores >300:
          scores = 300 
    else:
        print("學生沒有加分")
    
    print(f"學生總分為{scores:.0f}")
else:
    print("學生分數不可以大於300分")
print("程式結束")
```
